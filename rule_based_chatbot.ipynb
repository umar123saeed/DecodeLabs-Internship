## What is a Rule-Based Chatbot?
A Rule-Based Chatbot responds to user input using a **predefined set of rules** (a dictionary of question-answer pairs). It does not learn from data — instead, it matches the user's input to a known pattern and returns the corresponding response.

## How This Chatbot Works
1. User types a message
2. Input is cleaned (lowercased + stripped)
3. Cleaned input is matched against the knowledge base (dictionary)
4. If matched → returns the predefined response
5. If not matched → returns a fallback message

---

## Step 1: Knowledge Base
This is the brain of our chatbot — a Python dictionary where:
- **Keys** = expected user inputs (questions/keywords)
- **Values** = bot's responses

```python
# Knowledge Base — stores all question-answer pairs
# Key: what the user might type
# Value: what the bot will reply
responses = {
    'hello'         : 'Hi there! How can I help you?',
    'hi'            : 'Hello! What can I do for you?',
    'how are you'   : 'I am doing great! Thanks for asking.',
    'what is ai'    : 'AI is Artificial Intelligence — machines that think like humans!',
    'your name'     : 'I am DecodeBot, your AI assistant!',
    'help'          : 'I can answer questions about AI. Just ask me anything!',
    'what is ml'    : 'ML is Machine Learning — AI that learns from data!',
    'who made you'  : 'I was built by Umar Saeed Jan at DecodeLabs!',
    'bye'           : 'Goodbye! Have a great day!',
    'thanks'        : 'You are welcome! Anything else?'
}

print(f"Knowledge base loaded successfully!")
print(f"Total rules defined: {len(responses)}")
print(f"Bot can respond to: {list(responses.keys())}")
```

## Step 2: Response Function
This function handles:
- **Input sanitization** — converts to lowercase and removes extra spaces
- **Exit detection** — checks if user wants to quit
- **Dictionary lookup** — matches input to knowledge base
- **Fallback response** — handles unknown inputs gracefully

```python
def get_response(user_input):
    """
    Processes user input and returns an appropriate bot response.
    
    Args:
        user_input (str): Raw text typed by the user
    
    Returns:
        str: Bot's response message
        None: If user wants to exit
    """
    
    # Sanitize input — convert to lowercase and remove leading/trailing spaces
    # This ensures 'Hello', 'HELLO', 'hello' all match the same key
    clean_input = user_input.lower().strip()
    
    # Check if user wants to exit the chatbot
    if clean_input == 'exit':
        return None
    
    # Look up the cleaned input in the knowledge base dictionary
    # If not found, return a default fallback message
    reply = responses.get(
        clean_input,
        "I do not understand that yet. Try asking something else!"
    )
    
    return reply

print("Response function defined successfully!")
```

## Step 3: Test the Response Function
Let's test our function with some sample inputs before running the full chatbot.

```python
# Test cases to verify the chatbot works correctly
test_inputs = [
    'hello',           # Should match exactly
    'HELLO',           # Should still match (after lowercase conversion)
    'what is ai',      # Should return AI definition
    'what is ml',      # Should return ML definition
    'who made you',    # Should return creator info
    'random question'  # Should return fallback message
]

print("Testing chatbot responses:")
print("=" * 45)

for test in test_inputs:
    response = get_response(test)
    print(f"Input : {test}")
    print(f"Output: {response}")
    print("-" * 45)
```

## Step 4: Run the Full Chatbot
Now we run the complete interactive chatbot loop.

> **Note:** Type `exit` to stop the chatbot.

```python
# Display welcome banner
print("=" * 40)
print("   Welcome to DecodeBot! 🤖")
print("   Type 'exit' to quit")
print("=" * 40)

# Main chatbot loop — keeps running until user types 'exit'
while True:
    
    # Take input from user
    user_input = input("\nYou: ")
    
    # Get bot response
    reply = get_response(user_input)
    
    # If reply is None, user wants to exit — break the loop
    if reply is None:
        print("Bot: Goodbye! See you next time!")
        break
    
    # Print the bot's response
    print(f"Bot: {reply}")
```

**Key Concept:** Rule-based chatbots are simple but effective for narrow, well-defined use cases. They are the foundation before moving to ML-based NLP chatbots.
